```python
def rename_state(state_name):
    words = state_name.split()  # Split the state name into words
    updated_words = [word.lower() if word.lower() == 'and' else word.capitalize() for word in words]
    return ' '.join(updated_words)

# Apply the renaming function to the 'State/UT' column
new_data['State/UT'] = new_data['State/UT'].apply(rename_state)
```
